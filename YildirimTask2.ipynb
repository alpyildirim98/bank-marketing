```python
# It seems we have a lot of unknowns in some columns

# Let's confirm that
unknown_rows = df[(df['education'] == 'unknown') | (df['job'] == 'unknown') | (df['marital'] == 'unknown') |
 (df['default'] == 'unknown') | (df['housing'] == 'unknown') | (df['loan'] == 'unknown')]

total_count = len(unknown_rows)

print("Total number of observations with at least one 'unknown' category:", total_count)
```

We have 10700 unknowns in total. Completely discarding them is one approach, however this deletes almost 26% of the data, which would reduce the explanotary power of the dataset.

Before doing that, we should see whether observations with unknown variables affect the result of the outcome y.

```python
categorical_columns_2 = ['job', 'marital', 'education', 'default', 'housing', 'loan','y']
for column in categorical_columns_2:
    plt.figure(figsize=(10,4))
    #Returns counts of unique values for each outcome for each feature.
    pos_counts = df.loc[df.y.values == 'yes', column].value_counts()
    neg_counts = df.loc[df.y.values == 'no', column].value_counts()

    all_counts = list(set(list(pos_counts.index) + list(neg_counts.index)))

    #Counts of how often each outcome was recorded.
    freq_pos = (df.y.values == 'yes').sum()
    freq_neg = (df.y.values == 'no').sum()

    pos_counts = pos_counts.to_dict()
    neg_counts = neg_counts.to_dict()

    all_index = list(all_counts)
    all_counts = [pos_counts.get(k, 0) / freq_pos - neg_counts.get(k, 0) / freq_neg for k in all_counts]

    sns.barplot(x = all_counts, y=all_index)
    plt.title(column)
    plt.tight_layout()
```

Above, a positive value indicates clients who are more likely to subscribe according to their category, and negative value vice versa. It seems that unknowns in default may significantly impact the results.

One could argue that unknowns in the default are the type of clients who are not willing to share that information, and this preference might be a category by itself. Therefore, we are going to keep the unknown answers in default.

```python
#Creating train and test splits
from sklearn.model_selection import train_test_split

X = df_dummies.drop("y_yes", axis=1)
y = df_dummies["y_yes"]

# We stratify the sets to split in a way that preserves the same proportions of y_yes as observed in the original dataset.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
print('Original:', (df_dummies.y_yes).mean(), 'Train:', y_train.mean(), 'Test:', y_test.mean())
```

```python
del(X_train)
del(y_train)
del(X_test)
del(y_test)
```

Now, we will train the model using Logistic Regression and get the ROC AUC score

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


# Instantiate algorithm
model = LogisticRegression()
scaler = StandardScaler()

# Create training and test samples
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75, stratify = y, random_state=42)
# Scale X data
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

# Train model
trained_model = model.fit(X_train, y_train)

# Generate predicted probabilities on test sample
y_pred_proba = trained_model.predict_proba(X_test)[:, 1]

# Compute ROC AUC score
auc = roc_auc_score(y_true=y_test, y_score=y_pred_proba)
print(f"ROC AUC = {auc:.3f}")
```

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold
model = LogisticRegression(max_iter = 1000)
scaler = StandardScaler()

scaler.fit(X)
X_scaled = scaler.transform(X)

k = 5  # number of folds
cv = StratifiedKFold(n_splits=k, shuffle=True, random_state=42)

scores = cross_val_score(model, X_scaled, y, cv=cv, scoring='roc_auc')
mean_auc = scores.mean()
std_auc = scores.std()
print(f"Mean ROC AUC = {mean_auc:.3f}")
print(f"Standard Deviation of ROC AUC = {std_auc:.3f}")
```

Our Mean ROC AUC value as the result of 5-fold cross validations is 0.787, with a standard deviation of 0.006. The very low standard deviation of our results show that the results are robust, and the model is probably not overfitted.

A ROC AUC value of 0.787 indicates that the model has achieved a reasonably good level of discrimination between the classes.

Consumer price index and number of employees are also important factors with high explanatory power in the model.
